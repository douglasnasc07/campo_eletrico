### Campo elétrico de uma carga puntiforme:


$$
E = k\frac{Q}{r^2}
$$

$$
k = \frac{1}{4\pi\varepsilon_0}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

k = 9*np.exp(9)  # Constante eletrostática no vácuo
Q = 1*np.exp(-9)  # Carga da partícula (C)

r = np.linspace(1*np.exp(-5), 1*np.exp(-1), 1000)

E = (k*Q)/r**2
```

```python
plt.plot(r, E)
plt.xlabel('r')
plt.ylabel('E(N/C)')
plt.title('Campo Elétrico de uma Carga Pontual')
plt.show()
```

```python
Q = 1*np.exp(-9)


x, y = np.meshgrid(np.linspace(-0.5, 0.5, 50), np.linspace(-0.5, 0.5, 50))

r = np.sqrt(x**2 + y**2)
Ex = k*Q*x/r**3
Ey = k*Q*y/r**3
E = np.sqrt(Ex**2 + Ey**2)

plt.quiver(x, y, Ex, Ey)
plt.imshow(E,cmap='rainbow', extent=[-0.5, 0.5, -0.5, 0.5])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo Elétrico de uma Carga Pontual')
plt.show()
```

```python
Q = -1*np.exp(-9)

x, y = np.meshgrid(np.linspace(-0.5, 0.5, 50), np.linspace(-0.5, 0.5, 50))

r = np.sqrt(x**2 + y**2)
Ex = k*Q*x/r**3
Ey = k*Q*y/r**3
E = np.sqrt(Ex**2 + Ey**2)

plt.quiver(x, y, Ex, Ey)
plt.imshow(E,cmap='rainbow', extent=[-0.5, 0.5, -0.5, 0.5])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo Elétrico de uma Carga Pontual')
plt.show()
```

### Campo elétrico de uma esfera condutora:


$$
E = k\frac{Q}{R^2}
$$


```python
Q = 1 * 10**-9  # Carga da esfera (C)
R = 0.10  # raio da esfera (m)

r = np.linspace(1e-5, 1.0, 1000)

E = np.zeros_like(r)  # Inicializa um array de zeros com o mesmo tamanho de r
E[r > R] = (k * Q) / r[r > R]**2

plt.plot(r, E)
plt.xlabel('Distância (m)')
plt.ylabel('Campo Elétrico (N/C)')
plt.title('Campo Elétrico de uma esfera condutora carregada')
plt.show()
```

```python
Q = 1 * 10**-9  # Carga da esfera (C)
R = 0.10  # raio da esfera (m)

x, y = np.meshgrid(np.linspace(-0.5, 0.5, 50), np.linspace(-0.5, 0.5, 50))

r = np.sqrt(x**2 + y**2)

Ex = np.zeros_like(r)
Ey = np.zeros_like(r)

fora_esfera = r > R
Ex[fora_esfera] = k * Q * x[fora_esfera] / r[fora_esfera]**3
Ey[fora_esfera] = k * Q * y[fora_esfera] / r[fora_esfera]**3

E = np.sqrt(Ex**2 + Ey**2)

plt.quiver(x, y, Ex, Ey)
plt.imshow(E,cmap='rainbow', extent=[-0.5, 0.5, -0.5, 0.5])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo Elétrico de uma esfera condutora carregada')
plt.axis('equal')
plt.show()
```

### Campo elétrico de uma esfera isolante com densidade de carga volumétrica constante:

$$
E = k\frac{Q}{r^2} \hspace{0.5cm} (r > R)
$$

$$
E = kQ\frac{r}{R^3} \hspace{0.5cm} (r < R)
$$

```python
Q = 1 * 10**-9  # Carga da esfera (C)
R = 0.1 #raio da esfera (m)
r = np.linspace(0, 1.0, 100)

E = np.zeros_like(r)

fora_esfera = r > R
E[fora_esfera] = k * Q  / r[fora_esfera]**2

dentro_esfera = r < R
E[dentro_esfera] = k * Q * r[dentro_esfera]  / R**3

plt.plot(r, E)
plt.xlabel('Distância (m)')
plt.ylabel('Campo Elétrico (N/C)')
plt.title('Campo Elétrico de uma Esfera Isolante Carregada')
plt.show()
```

### Campo elétrico de uma casca esférica isolante com densidade de carga volumétrica constante:

$$
E = 0 \hspace{0.5cm} (r < R_i)
$$

$$
E = k\frac{Q}{r^2} \frac{r^3 - R_i^3}{R_e^3 - R_i^3} \hspace{0.5cm} (R_i < r < R_e)
$$

$$
E = k\frac{Q}{r^2} \hspace{0.5cm} ( r > R_e)
$$

```python
Q = 1 * 10**-9  # Carga da esfera (C)
Ri = 0.3 #raio da esfera (m)
Re = 0.5 #raio externo

r = np.linspace(1*np.exp(-4), 2.0, 100)

E = np.zeros_like(r)

fora_esfera = r >= Re
E[fora_esfera] = k * Q  / r[fora_esfera]**2

casca = r < Re
E[casca] = ((k*Q)/r[casca]**2)*((r[casca]**3 - Ri**3)/(Re**3 - Ri**3))

dentro_casca = r < Ri
E[dentro_casca] = 0

plt.plot(r, E)
plt.xlabel('Distância (m)')
plt.ylabel('Campo Elétrico (N/C)')
plt.title('Campo Elétrico de uma casca esférica isolante carregada')
plt.show()
```

```python
Q = 1 * 10**-9  # Carga da esfera (C)
Ri = 0.3 #raio interno da esfera (m)
Re = 0.5 #raio externo (m)

x, y = np.meshgrid(np.linspace(-1, 1, 50), np.linspace(-1, 1, 50))

r = np.sqrt(x**2 + y**2)

Ex = np.zeros_like(r)
Ey = np.zeros_like(r)

fora_esfera = r >= Re
Ex[fora_esfera] = k * Q * x[fora_esfera] / r[fora_esfera]**3
Ey[fora_esfera] = k * Q * y[fora_esfera] / r[fora_esfera]**3

casca = r < Re
Ex[casca] = ((k*Q)/r[casca]**2)*((r[casca]**3 - Ri**3)/(Re**3 - Ri**3))*x[casca]
Ey[casca] = ((k*Q)/r[casca]**2)*((r[casca]**3 - Ri**3)/(Re**3 - Ri**3))*x[casca]

dentro_casca = r < Ri
Ex[dentro_casca] = 0
Ey[dentro_casca] = 0

E = np.sqrt(Ex**2 + Ey**2)

plt.quiver(x, y, Ex, Ey)
plt.imshow(E,cmap='rainbow', extent=[-1, 1, -1, 1])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Campo Elétrico de uma casca esférica isolante carregada')
plt.axis('equal')
plt.show()
```
